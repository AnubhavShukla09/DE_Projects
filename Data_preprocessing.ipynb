**BATTING SUMMARY**

```python
with open("t20_json_files/t20_wc_batting_summary.json") as f:
    data = json.load(f)
data
```

```python
len(data)
```

```python
data[0]["battingSummary"]
```

```python
data[1]["battingSummary"]
```

```python
total_rows = []

for i in data:
    total_rows.extend(i["battingSummary"])

df_batting_summary = pd.DataFrame(total_rows)
df_batting_summary
```

```python
def out_func(word):
    if len(word)>0:
        return "Out"
    else:
        return "Not Out"
```

```python
df_batting_summary["dismissal"] = df_batting_summary["dismissal"].apply(out_func)
#df_batting_summary["dismissal"].apply(lambda x:"Out" if len(x)>0 else "Not Out")
```

```python
df_batting_summary
```

```python
df_batting_summary.rename(columns = {
    "dismissal" : "Out/Not Out"
},inplace=True)
```

```python
df_batting_summary["batsmanName"].unique
```

```python
pd.set_option("display.max_rows",20)
```

```python
df_batting_summary['batsmanName'] = df_batting_summary['batsmanName'].apply(lambda x: x.replace('', ''))
df_batting_summary['batsmanName'] = df_batting_summary['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting_summary
```

```python
df_match_results
```

```python
match_dict = {}

for index,row in df_match_results.iterrows():
    key1 = row["team1"]+" Vs "+row["team2"]
    key2 = row["team2"]+" Vs "+row["team1"]

    match_dict[key1] = row["match_id"]
    match_dict[key2] = row["match_id"]

match_dict
```

```python
df_batting_summary["match"].apply(lambda x:match_dict[x])
```

```python
df_batting_summary["match_id"] = df_batting_summary["match"].apply(lambda x:match_dict[x])
df_batting_summary
```

```python
df_batting_summary.to_csv("t20_csv_files/fact_batting_summary.csv",index=False)
```
