```python
import pandas as pd
import numpy as np
```

**CANDIDATES**

```python
candidates_df = pd.read_csv('source_files/candidates_with_phase.csv')
```

```python
candidates_df
```

```python
candidates_df['State'] = candidates_df['State'].apply(lambda x:x.upper())
candidates_df['Constituency'] = candidates_df['Constituency'].apply(lambda x:x.upper())
candidates_df['Party'] = candidates_df['Party'].apply(lambda x:x.upper())
candidates_df['Candidate Name'] = candidates_df['Candidate Name'].apply(lambda x:x.upper())
candidates_df['Gender'] = candidates_df['Gender'].apply(lambda x:x.upper())
```

```python
candidates_df
```

```python
candidates_df.to_csv('datasets/Candidates.csv',index=False)
```

**RESULTS**

```python
results_df = pd.read_csv('source_files/election_results.csv')
```

```python
results_df.rename(columns={'PC Name': 'Constituency'},inplace=True)
```

```python
results_df.rename(columns={'PC N0': 'ConstituencyKey'},inplace=True)
```

```python
results_df
```

```python
results_df['State'] = results_df['State'].apply(lambda x:x.upper())
results_df['Constituency'] = results_df['Constituency'].apply(lambda x:x.upper())
results_df['Party'] = results_df['Party'].apply(lambda x:x.upper())
results_df['Candidate'] = results_df['Candidate'].apply(lambda x:x.upper())
```

```python
results_df
```

```python
results_df["EVM Votes"].replace("-",0,inplace=True)
results_df["Postal Votes"].replace("-",0,inplace=True)
results_df["Total Votes"].replace("-",0,inplace=True)
results_df["Vote Share"].replace("-",0,inplace=True)
results_df[results_df["EVM Votes"]==0]
```

```python
results_df.to_csv('datasets/Results.csv',index = False)
```

**WINNERS**

```python
winners_df = pd.read_csv('source_files/election_winners.csv')
```

```python
winners_df
```

```python
winners_df[winners_df.Status!='Result Declared']
```

```python
winners_df["Trailing Candidate"].fillna("", inplace=True)
winners_df["Trailing Party"].fillna("", inplace=True)
winners_df["Margin"].replace("-",0,inplace=True)
winners_df[winners_df.Constituency == "Surat"]
```

```python
winners_df['Constituency'] = winners_df['Constituency'].apply(lambda x:x.upper())
winners_df['Leading Candidate'] = winners_df['Leading Candidate'].apply(lambda x:x.upper())
winners_df['Leading Party'] = winners_df['Leading Party'].apply(lambda x:x.upper())
winners_df['Trailing Candidate'] = winners_df['Trailing Candidate'].apply(lambda x:x.upper())
winners_df['Trailing Party'] = winners_df['Trailing Party'].apply(lambda x:x.upper())
```

```python
winners_df
```

```python
winners_df.rename(columns = {'Const. No.':'ConstituencyKey'},inplace = True)
```

```python
winners_df
```

```python
winners_df.to_csv('datasets/Winners.csv',index=False)
```
